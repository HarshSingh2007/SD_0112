```python
name=str(input("Enter your name:"+" "))
course=str(input("Enter your course:"+" "))
branch=str(input("Enter your branch:"+" "))
rollNumber=int(input("Enter your Roll number:"+" "))
print("-----Please enter the marks for following subjects out of 100-----\n")
m1=int(input("Enter marks of Python:"+" "))
m2=int(input("Enter marks of COA:"+" "))
m3=int(input("Enter marks of Maths-4:"+" "))
m4=int(input("Enter marks of DSTL:"+" "))
m5=int(input("Enter marks of Universal Values:"+" "))
sum=m1+m2+m3+m4+m5
print("-----Total marks obtained is------"+" "+str(sum))
per=sum/5
print("-----Percentage is=----"+" "  +str(per))

if per>=90:
  print("-----Grade A-----")
elif per>=80:
  print("-----Grade B------")
elif per>=70:
  print("-----Grade C-----")
elif per>=60:
  print("-----Grade D------")

```
