```python
import time
from IPython.display import clear_output
from datetime import datetime
while True:
  clear_output(wait=True)
  current_time=datetime.now().strftime("%I:%M:%S %p")
  print("    DIGITAL CLOCK")
  print(current_time)
  time.sleep(1)
```
